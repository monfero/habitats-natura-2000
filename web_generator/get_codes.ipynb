```python
# Cada concellos pertence a unha área diferente 
# Este script permite crear unha listaxa que asigan a cada concello o seu código correspondente
# Esta información utilízase á hora de xerar as páxinas web
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv('../datasets/completos/habitats_aerox.csv',sep=';')
df2 = pd.read_csv('../datasets/completos/habitats_laats.csv',sep=';')
df3 = pd.read_csv('../datasets/completos/habitats_poligonales.csv',sep=';')
```

```python
df1 = df1[['concello','ua_tesel_1']]
df1['ua_tesel_1'] = df1.ua_tesel_1.apply(lambda x: x.split('.')[0])
```

```python
df2 = df2[['concello','ua_tesel_1']]
df2['ua_tesel_1'] = df2.ua_tesel_1.apply(lambda x: x.split('.')[0])
```

```python
df3 = df3[['concello','ua_tesel_1']]
df3['ua_tesel_1'] = df3.ua_tesel_1.apply(lambda x: x.split('.')[0])
```

```python
dftotal = pd.concat([df1,df2,df3])
dftotal = dftotal.drop_duplicates()
```
